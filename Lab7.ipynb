```
import nltk
from nltk.corpus import stopwords
import string

# Bổ sung tải đầy đủ các gói tách từ để tránh lỗi hoàn toàn trên Colab
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True) # Sửa lỗi hệ thống tách từ mới của NLTK
nltk.download('stopwords', quiet=True)

print("## Bài 4: Loại bỏ stopword khỏi văn bản")
text = "This is a sample sentence showing how to remove stopwords from a given text."

stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

# Tách từ
words = nltk.word_tokenize(text)

# Lọc: Đưa về chữ thường .lower() để so sánh chính xác với stop_words
filtered_text = [w for w in words if w.lower() not in stop_words and w not in punctuation]

print("Văn bản gốc:", words)
print("Sau khi lọc sạch stopword & dấu câu:", filtered_text)
```

```
from nltk.corpus import wordnet

print("## Bài 6: Tra cứu WordNet")
word = "computer"
syns = wordnet.synsets(word)

if syns:
    print(f"Từ khóa: '{word}'")
    print("Định nghĩa:", syns[0].definition())
    print("Ví dụ minh họa:", syns[0].examples())
else:
    print("Không tìm thấy dữ liệu.")
```

```
from nltk.corpus import wordnet

print("## Bài 7: Tìm từ đồng nghĩa & trái nghĩa")
word = "beautiful"
synonyms = []
antonyms = []

for syn in wordnet.synsets(word):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print(f"Từ gốc: {word}")
print("- Từ đồng nghĩa (Synonyms):", set(synonyms))
print("- Từ trái nghĩa (Antonyms):", set(antonyms))
```

```
import nltk
import re

# Tải gói dữ liệu tagsets
nltk.download('tagsets', quiet=True)

print("## Bài 8: Tổng quan và chi tiết bộ tag bằng Biểu thức chính quy (RegEx)")
print("-" * 60)

# Tải file dữ liệu gốc của bộ tag Penn Treebank
try:
    tagdict = nltk.data.load('help/tagsets/upenn_tagset.pickle')
except LookupError:
    # Nếu không tải được bằng pickle, ta dùng định dạng văn bản mặc định
    tagdict = nltk.corpus.repo_txt if hasattr(nltk.corpus, 'repo_txt') else {}

# 1. Chi tiết một tag cụ thể (Ví dụ: 'NN')
target_tag = 'NN'
if target_tag in tagdict:
    print(f"[1. Chi tiết của tag cụ thể '{target_tag}']: ")
    print(f" - Định nghĩa: {tagdict[target_tag][0]}")
    print(f" - Ví dụ: {tagdict[target_tag][1]}")
print("-" * 60)

# 2. Sử dụng biểu thức chính quy để tìm các bộ tag liên quan
# Ví dụ: Tìm tất cả các tag bắt đầu bằng 'NN' (Danh từ) -> RegEx: ^NN
regex_pattern = r'^NN'

print(f"[2. Các tag liên quan khớp với biểu thức chính quy '{regex_pattern}']:")
found = False
for tag, (definition, examples) in tagdict.items():
    if re.match(regex_pattern, tag):
        print(f" * Thẻ {tag:<5} | Ý nghĩa: {definition}")
        print(f"   Ví dụ: {examples}")
        found = True

if not found:
    print("Không tìm thấy thẻ nào khớp với RegEx.")
print("-" * 60)

# 3. Một ví dụ khác với Động từ: Tìm tất cả các tag bắt đầu bằng 'VB' -> RegEx: ^VB
regex_pattern_vb = r'^VB'
print(f"[3. Các tag liên quan khớp với biểu thức chính quy '{regex_pattern_vb}']:")
for tag, (definition, examples) in tagdict.items():
    if re.match(regex_pattern_vb, tag):
        print(f" * Thẻ {tag:<5} | Ý nghĩa: {definition}")
```

```
from nltk.corpus import wordnet

print("## Bài 9: Độ tương đồng giữa 2 danh từ")
# Lấy nghĩa (synset) đầu tiên của cụm từ
noun1 = wordnet.synset('dog.n.01')
noun2 = wordnet.synset('cat.n.01')

# Tính toán độ tương đồng Path Similarity (giá trị từ 0 đến 1)
similarity = noun1.path_similarity(noun2)
print(f"Độ tương đồng giữa 'dog' (chó) và 'cat' (mèo): {similarity:.4f}")
```

```
from nltk.corpus import wordnet

print("## Bài 10: Độ tương đồng giữa 2 động từ")
verb1 = wordnet.synset('run.v.01')
verb2 = wordnet.synset('walk.v.01')

similarity = verb1.path_similarity(verb2)
print(f"Độ tương đồng giữa 'run' (chạy) và 'walk' (đi bộ): {similarity:.4f}")
```

```
import random
from nltk.corpus import names

print("## Bài 12: Trộn ngẫu nhiên tên kèm nhãn giới tính")
# Tạo list tuple (Tên, Nhãn)
labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Trộn ngẫu nhiên dữ liệu (Dùng seed để kết quả không đổi mỗi lần chạy thử)
random.seed(42)
random.shuffle(labeled_names)

print("15 cặp tên và nhãn ngẫu nhiên sau khi trộn:")
for name, gender in labeled_names[:15]:
    print(f" - Tên: {name:<12} | Nhãn: {gender}")
```

```
from nltk.corpus import names

print("## Bài 13: Trích xuất ký tự cuối cùng của tên làm đặc trưng (Features)")
labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Tạo mảng mới chỉ lấy (Ký tự cuối, Nhãn)
last_letter_features = [(name[-1].lower(), gender) for (name, gender) in labeled_names]

print("15 phần tử đầu tiên trong mảng đặc trưng mới:")
for feature in last_letter_features[:15]:
    print(feature)
```
