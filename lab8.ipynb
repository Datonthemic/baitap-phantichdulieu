```
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "This is a simple example showing how to remove stopwords using NLTK."

words = word_tokenize(text)

filtered = [word for word in words if word.lower() not in stopwords.words('english')]

print("Trước:")
print(words)

print("\nSau khi loại stopwords:")
print(filtered)
```

```
from nltk.corpus import stopwords

words = ['I','love','to','study','Python','with','NLTK']

result = []

for word in words:
    if word.lower() not in stopwords.words('english'):
        result.append(word)

print(result)
```

```
from nltk.corpus import wordnet

word = input("Nhập từ: ")

for syn in wordnet.synsets(word):
    print("Tên:", syn.name())
    print("Định nghĩa:", syn.definition())
    print("Ví dụ:", syn.examples())
    print()
```

```
from nltk.corpus import wordnet

word = input("Nhập từ: ")

synonyms = set()
antonyms = set()

for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("\nĐồng nghĩa:")
print(synonyms)

print("\nTrái nghĩa:")
print(antonyms)
```

```
from nltk.corpus import wordnet

n1 = input("Danh từ 1: ")
n2 = input("Danh từ 2: ")

s1 = wordnet.synsets(n1, pos=wordnet.NOUN)
s2 = wordnet.synsets(n2, pos=wordnet.NOUN)

if s1 and s2:
    print(s1[0].wup_similarity(s2[0]))
else:
    print("Không tìm thấy.")
```

```
from nltk.corpus import names

male = names.words('male.txt')
female = names.words('female.txt')

print("Số tên nam:", len(male))
print("Số tên nữ:", len(female))

print("\n10 tên nam đầu:")
print(male[:10])

print("\n10 tên nữ đầu:")
print(female[:10])
```

```
from nltk.corpus import names
import random

labeled_names = [(name, 'male') for name in names.words('male.txt')]
labeled_names += [(name, 'female') for name in names.words('female.txt')]

random.shuffle(labeled_names)

for item in labeled_names[:15]:
    print(item)
```

```
from nltk.corpus import names

features = []

for name in names.words('male.txt'):
    features.append((name[-1], 'male'))

for name in names.words('female.txt'):
    features.append((name[-1], 'female'))

for item in features[:20]:
    print(item)
```
